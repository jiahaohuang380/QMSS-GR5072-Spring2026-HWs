However, the Jupyter Notebook (.ipynb) code looked "foreign" and messy. This is because an .ipynb file is actually a JSON file. It contains not only my code and text but also a lot of metadata, cell information, and formatting code that Jupyter needs to render the notebook in a browser. That is why it looks like a complex structure of brackets and quotes rather than just simple code.
