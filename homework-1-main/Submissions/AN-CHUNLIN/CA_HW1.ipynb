## 1. Project description <a class="anchor" id="bullet1"></a>

This project utilizes convoluted neural networks to train an algorithm that (with enough image input of boss) recognizes the *boss*'s face as distinct from *other* faces. When the boss appears in the web camera, the program fetches the image and automatically displays a PyCharm screenshot to hide other tabs on the screen. (I think it's hilarious that the creator probably collected tons of images of his boss to train the model)

* ### Some files included in this repository:
  * `data` folder: intended storage for both images that are labeled as 'boss' or 'other'.
  * `boss_input.py`: for loading, reading, and extracting images of boss.
  * `boss_train.py`: for training, validating, and testing a CNN on boss image classification.
