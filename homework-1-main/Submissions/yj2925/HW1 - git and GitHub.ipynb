## 1. Project Overview: Microsoft Qlib <a class="anchor" id="bullet1"></a>
### My Struggles with Financial Data
As a student interested in **Economic trends and Market behavior**, I used to spend countless hours searching for clean, structured financial data. My experience was often a "nightmare":

* **Data Fragmentation**: Market data is scattered across different platforms with inconsistent formats.

* **High Barrier to Entry**: Converting raw market data into something an AI model can understand requires complex preprocessing.

* **"Headache" of Integration**: I often got a headache trying to align timestamps from different economic indicators.

I discovered **Microsoft Qlib** (https://github.com/microsoft/qlib?tab=readme-ov-file), which is an AI-oriented quantitative investment platform. It solves my problems by providing a full-stack infrastructure for data prep and model training.

## 2. Why Qlib is a Game Changer for Researchers <a class="anchor" id="bullet2"></a>
This project is fascinating because it bridges the gap between traditional _finance and modern AI_. It includes:

* **Built-in Data Loader**: It can automatically download and prepare historical market data.

* **Rich Model Zoo**: It features State-of-the-Art (SOTA) AI models like LightGBM and GAT.

* **Performance Analysis**: It provides tools to visualize how a trading strategy would have performed in the past.

### 3. Reflection on Code Visibility<a class="anchor" id="bullet3"></a>

To answer the Qustions 6, when reviewing the commit history, **the code for the Python script (.py) looked very familiar** because it is a simple plain-text file that only contains the exact lines of code I wrote. However, **the code for the Jupyter Notebook (.ipynb) appeared foreign and complex**. This is because a Notebook file is actually **a JSON-formatted document**. Beyond my own code and text, it contains extensive **metadata**, such as cell types, execution counts, and formatting information. This extra structure makes the raw file difficult for a human to read directly in a version control history, even though it looks clean within the Jupyter interface.
